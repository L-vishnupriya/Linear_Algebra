**Matrix and Vector Initialization :**

* P is a matrix whose columns are basis vectors.

* D is a diagonal matrix with scaling factors.

* P_inv is the inverse of P.

* x is a vector.

* x_b is x represented in the new basis (using the inverse of P).

* scaled_x_b scales the coordinates in the new basis by the diagonal matrix D.

* x_new is the transformed vector in the original basis after scaling.

```python
# Define the matrices and vectors
P = np.array([[2, 1], [-1, 1]])
D = np.diag([3, 6])
P_inv = np.linalg.inv(P)
x = np.array([[2], [1]])
x_b = P_inv @ x
scaled_x_b = D @ x_b
x_new = P @ scaled_x_b
```
