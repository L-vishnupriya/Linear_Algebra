**Import Libraries :**

```python
import numpy as np
```

**Matrix A :**

```python
# Define the matrix A
A = np.array([[1, 3, 3], [-3, -5, -3], [3, 3, 1]])
print("Matrix A is:",A)
```

```python
# Calculate eigenvalues and eigenvectors
eigenvalues, eigenvectors = np.linalg.eig(A)
```

**Matrix P :**

```python
# Form matrix P of eigenvectors
P = eigenvectors
print("Matrix P (Eigenvectors):\n", P)
```

**Matrix D :**

```python
# Form matrix D of eigenvalues
D = np.diag(eigenvalues)
print("\nMatrix D (Eigenvalues):\n", D)
```

**Matrix P^-1 (Inverse of P):**

```python
# Calculate P inverse (P^-1)
P_inv = np.linalg.inv(P)
print("\nMatrix P^-1 (Inverse of P):\n", P_inv)
```

```python
# Multiply PDP^-1
result = np.dot(P, np.dot(D, P_inv))

# Check if result matches A (up to numerical precision)
if np.allclose(result, A):
    print("PDP^-1 matches A.")
else:
    print("PDP^-1 does not match A.")

# To check if columns of P are independent
# Compute the determinant of P
det_P = np.linalg.det(P)
print('-------------------')
# If determinant is non-zero, columns are independent
if det_P != 0:
    print("Columns of P are independent.")
else:
    print("Columns of P are dependent.")
```

**Conclusion :**

Thus the , Column of P are independent
