# 1.Formulate the equations

```python
# 1. Formulate the equations
# x1 = x3 + 1/2 * x4
# x2 = 1/3 * x1
# x3 = 1/3 * x1 + 1/2 * x2 + 1/2 * x4
# x4 = 1/3 * x1 + 1/2 * x2
```

# 2.Create Matrix A:
                In this matrix:
* Each row represents a page.
* Each column represents links from a page to others.
* The values in the matrix represent the proportion of links from one page to another.

```python
import numpy as np

A = np.array([
    [0, 0, 1, 1/2],
    [1/3, 0, 0, 0],
    [1/3, 1/2, 0, 1/2],
    [1/3, 1/2, 0, 0]
])
```

# 3.Find eigenvalues and eigenvectors:
                            

```python
eigenvalues, eigenvectors = np.linalg.eig(A.T)
```

* np.linalg.eig() calculates the eigenvalues and eigenvectors of a square matrix.
* T is the transpose of the matrix, which is used because the PageRank algorithm traditionally works on the transpose of the adjacency matrix.

# 4.Find the dominant eigenvalue

```python
dominant_index = np.argmax(np.abs(eigenvalues))
```

* np.argmax() returns the index of the maximum value.
* np.abs() calculates the absolute value of each element.

# 5.Get the corresponding eigenvector

```python
eigenvector = np.real(eigenvectors[:, dominant_index])
```

* eigenvectors[:, dominant_index] selects the eigenvector corresponding to the dominant eigenvalue.
* np.real() takes the real part of the eigenvector.

# 6.Rescale the eigenvector

```python
pagerank = eigenvector / np.sum(eigenvector)
```

# 7.Page Importance Display

```python
print("Page Importance:")
for i, importance in enumerate(pagerank, start=1):
    print(f"Page {i}: {importance}")
```

# 8.Page Ranking

```python
page_ranking = np.argsort(pagerank)[::-1]
print("\nPage Ranking (Most important to least important):")
for rank, page in enumerate(page_ranking, start=1):
    print(f"Rank {rank}: Page {page+1}")
```
